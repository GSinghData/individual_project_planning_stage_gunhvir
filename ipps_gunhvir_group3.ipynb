**Data Description**
The above data has 196 unique observations. It contains 7 different variables named: experience (character), subscribe (logical), hashedEmail (character), played_hours (double), name (character), gender (character), and age (double). The experience variable is a description of how experienced each user is at the game of Minecraft. It is a list of an arbitrary size, with the possible values being: Beginner, Amateur, Regular, Veteran, and Pro. The subscribe variable, which is either true or false, records whether or not each user is subscribed to newletters from the team performing the study. The hashedEmail variable represents each user's email used to sign up without giving out their personal information. It is unique for each user and consists of letters and numbers. The played_hours variable keeps track of how many hours each user has played on the server for. The name variable is simply the name of the user. The gender variable is a catergory that is to be filled in by each user. Because each user identifies themselves in a way that they feel best represents them, and this is not bound to a limited number of possibilities, the number of possibilities for this variable are infinite, but the values that have been observed so far include: Male, Female, Non-Binary, Agender, Two-Spirited, Prefer not to say, and Other. The final variable, age, simply records the age of each user in years. The value of this column can be either a number, or NA.

A potential issut that could arrive from this data is the presence of the NA value in the age column. If someone were using the data from this column, they would have to either exclude the column by setting na.rm = TRUE, or calculate the value by averaging other ages that were close to that value on the plot.



**Question**
The Broad Question that I will be addressing is: We would like to know which "kinds" of players are most likely to contribute large amounts of data so we can target those players in our recruiting efforts.

The specific question I will be asking is can the age of the user and the number of hours played predict the experience level of that user?

The data above will help me to answer this question as the dataset I read in contians both the age of the users and their experience level. 

**Methods and Plan**
In order to address the question of "Can age and the number of hours played predict a user's experience level?", I will use a k-Nearest Neighbors (k-NN) classification model. This method aligns well with the structure of the dataset, the type of outcome, and the sample size. 

I believe that this is the most suitable method because the outcome variable, experience level, is categorical, and the predictors are numeric. Because k-NN does not need a relationship between predictors and the outcome, it is suitable for this problem. Additionally, the number of observations, 196, is suitable for k-NN.

k-NN requires very few formal statistical assumptions. The main considerations are:

- Distance meaningfully reflects similarity: Users who are closer in age and in the number of hours played should plausibly have similar experience levels.
- Variables should be on comparable scales, otherwise one variable can dominate the distance calculation. Standardization will address this.
- No strong outliers, since k-NN predictions are sensitive to points that sit far from the main clusters.
- Unlike logistic regression, k-NN does not assume linearity, normality, or independence of residuals.

k-NN can perform poorly if:

- The predictors are not scaled (because distance becomes misleading).
- There is high noise or many irrelevant features (not a problem in this dataset).
- The value of k is poorly chosen. A small k can overfit, while a large k can oversmooth important class boundaries.
- Class imbalance exists. If one experience level dominates the dataset, k-NN may bias predictions toward that category.

I will evaluate the model using repeated cross-validation. Using tidymodels, I will tune the hyperparameter k across a range of values (e.g., k = 1 to 25) and compare performance based on:
- Accuracy
- Balanced accuracy (if classes are imbalanced)
- The chosen model will be the one that provides the highest cross-validated performance without signs of overfitting.

Before modelling, I will:

Split the data into a training set with a 75/25% split (75% training and 25% testing)
Apply a recipe that standardizes predictors using step_scale() and step_center(), since k-NN requires scaling, encodes the categorical outcome with step_dummy() if necessary (depending on the engine), use 10-fold cross-validation inside the training set to tune the hyperparameter k. After selecting the best k, I will fit the final model to the full training set and evaluate its performance on the testing set. This workflow ensures a fair and unbiased estimate of how well age and played_hours can predict a user’s experience level.
