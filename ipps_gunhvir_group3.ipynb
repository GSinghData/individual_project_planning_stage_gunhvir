**Data Description**
The above data has 196 unique observations. It contains 7 different variables named: experience (character), subscribe (logical), hashedEmail (character), played_hours (double), name (character), gender (character), and age (double). The experience variable is a description of how experienced each user is at the game of Minecraft. It is a list of an arbitrary size, with the possible values being: Beginner, Amateur, Regular, Veteran, and Pro. The subscribe variable, which is either true or false, records whether or not each user is subscribed to newletters from the team performing the study. The hashedEmail variable represents each user's email used to sign up without giving out their personal information. It is unique for each user and consists of letters and numbers. The played_hours variable keeps track of how many hours each user has played on the server for. The name variable is simply the name of the user. The gender variable is a catergory that is to be filled in by each user. Because each user identifies themselves in a way that they feel best represents them, and this is not bound to a limited number of possibilities, the number of possibilities for this variable are infinite, but the values that have been observed so far include: Male, Female, Non-Binary, Agender, Two-Spirited, Prefer not to say, and Other. The final variable, age, simply records the age of each user in years. The value of this column can be either a number, or NA.

A potential issut that could arrive from this data is the presence of the NA value in the age column. If someone were using the data from this column, they would have to either exclude the column by setting na.rm = TRUE, or calculate the value by averaging other ages that were close to that value on the plot.
